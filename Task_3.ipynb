**URL_1**

```python
# Import necessary libraries
from bs4 import BeautifulSoup
import requests

# Step 1: Define the URL of the static web page to scrape
url = "http://books.toscrape.com/"  # Replace with your desired static URL

# Step 2: Send a GET request to the webpage
response = requests.get(url)

# Step 3: Check if the request was successful
if response.status_code == 200:
    print("Successfully retrieved the webpage!\n")
    
    # Step 4: Parse the HTML content of the page
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # Step 5: Extract and display relevant information
    # (a) Extract all text from the page
    print("--- Page Text ---")
    page_text = soup.get_text()
    print(page_text[:500])  # Display the first 500 characters for brevity

    # (b) Extract all links from the page
    print("\n--- Links ---")
    links = soup.find_all('a')
    for link in links:
        href = link.get('href')
        if href:
            print(href)

    # (c) Extract all image sources from the page
    print("\n--- Image Sources ---")
    images = soup.find_all('img')
    for img in images:
        src = img.get('src')
        if src:
            print(src)

else:
    print(f"Failed to retrieve the webpage. Status code: {response.status_code}")
```

```python
with open("output.txt", "w", encoding="utf-8") as file:
    file.write("Page Text:\n")
    file.write(page_text)
    file.write("\n\nLinks:\n")
    for link in links:
        href = link.get('href')
        if href:
            file.write(href + "\n")
    file.write("\n\nImage Sources:\n")
    for img in images:
        src = img.get('src')
        if src:
            file.write(src + "\n")
print("Data saved to output.txt")
```

**URL_4**

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Corrected URL for the desktop version of Wikipedia
url = "https://en.wikipedia.org/wiki/List_of_largest_companies_in_the_United_States_by_revenue"

# Sending a GET request to the website
response = requests.get(url)

# Parsing the HTML content of the page
soup = BeautifulSoup(response.text, 'html.parser')

# Finding the table containing the data
table = soup.find('table', {'class': 'wikitable'})

if table:
    # Extracting table headers
    headers = [header.text.strip() for header in table.find_all('th')]

    # Extracting table rows
    rows = table.find_all('tr')

    # Extracting data from rows
    data = []
    for row in rows[1:]:  # Skip the header row
        cells = row.find_all(['td', 'th'])
        row_data = [cell.text.strip() for cell in cells]
        data.append(row_data)

    # Creating a DataFrame
    df = pd.DataFrame(data, columns=headers)

    # Saving data to an Excel file
    output_file = "largest_companies_USA.xlsx"
    df.to_excel(output_file, index=False)

    print(f"Data successfully scraped and saved to {output_file}")
else:
    print("Failed to find the table on the webpage. Please check the URL or class name.")
```
